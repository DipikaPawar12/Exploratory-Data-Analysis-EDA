**Drop All Null Columns & Rows whose all values are NULL**

---



```
df = data.dropna(how='all', axis=1)
df = df.dropna(how='all')
df
```

**Remove duplicate rows who have all NUll entries and all those rows who has to be unique value as column**

```
df = df.drop_duplicates()
df = df.drop_duplicates(subset=['UNITID','OPEID','OPEID6'])
df
```

**Check if any Column have same single value**

```
res=df
for col in df.columns:
    if len(df[col].unique()) == 1:
        res = res.drop([col],axis=1)
res
```
